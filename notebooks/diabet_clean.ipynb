# Penjelasan Label Encoding Dataset Diabetes

## Apa itu Label Encoding?
Label Encoding adalah teknik untuk mengubah data kategorikal (text) menjadi angka agar dapat diproses oleh algoritma machine learning.

---

## Kolom yang Di-Encode

### 1. GENDER (Jenis Kelamin)

**Mapping Encoding:**
| **Nilai Asli** | **Setelah Encoding** |
|----------------|----------------------|
| Female | 0 |
| Male | 1 |

**Contoh Transformasi:**
| **Sebelum** | **Sesudah** |
|-------------|-------------|
| Female | 0 |
| Male | 1 |
| Female | 0 |
| Male | 1 |

---

### 2. SMOKING_HISTORY (Riwayat Merokok)

**Mapping Encoding:**
| **Nilai Asli** | **Setelah Encoding** | **Keterangan** |
|----------------|----------------------|----------------|
| current | 0 | Perokok aktif |
| ever | 1 | Pernah merokok |
| former | 2 | Mantan perokok |
| never | 3 | Tidak pernah merokok |
| not current | 4 | Tidak merokok saat ini |

**Contoh Transformasi:**
| **Sebelum** | **Sesudah** |
|-------------|-------------|
| never | 3 |
| former | 2 |
| current | 0 |
| not current | 4 |
| ever | 1 |

---

## Kolom yang TIDAK Di-Encode

Kolom berikut **sudah dalam format numerik** (0 dan 1):

| **Kolom** | **Nilai** | **Arti** |
|-----------|----------|----------|
| hypertension | 0 | Tidak ada hipertensi |
| | 1 | Ada hipertensi |
| heart_disease | 0 | Tidak ada penyakit jantung |
| | 1 | Ada penyakit jantung |
| diabetes | 0 | Tidak diabetes |
| | 1 | Diabetes |

**Kolom numerik lainnya:** age, bmi, HbA1c_level, blood_glucose_level (sudah berupa angka).

---

## Perbandingan Sebelum dan Sesudah

### Sebelum Encoding:
| **gender** | **smoking_history** | **hypertension** | **heart_disease** |
|------------|---------------------|------------------|-------------------|
| Female | never | 0 | 0 |
| Male | former | 1 | 1 |
| Female | current | 0 | 0 |
| Male | not current | 1 | 0 |

### Sesudah Encoding:
| **gender** | **smoking_history** | **hypertension** | **heart_disease** |
|------------|---------------------|------------------|-------------------|
| 0 | 3 | 0 | 0 |
| 1 | 2 | 1 | 1 |
| 0 | 0 | 0 | 0 |
| 1 | 4 | 1 | 0 |

---

## Kenapa Perlu Encoding?

**❌ Tanpa Encoding:**
```python
model.fit(df[['gender', 'smoking_history']], df['diabetes'])
# ERROR: could not convert string to float: 'Female'
```

**✅ Dengan Encoding:**
```python
model.fit(df[['gender', 'smoking_history']], df['diabetes'])
# Berhasil! Model bisa dilatih
```

---

## Ringkasan

### Yang Di-Encode:
- **gender**: Female=0, Male=1
- **smoking_history**: current=0, ever=1, former=2, never=3, not current=4

### Yang Sudah Numerik (Tidak Di-Encode):
- **hypertension, heart_disease, diabetes**: Sudah 0/1
- **age, bmi, HbA1c_level, blood_glucose_level**: Sudah angka
